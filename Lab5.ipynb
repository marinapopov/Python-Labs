```python
dft = df
dft
```

```python
#separarea coloanelor dupa tipul de date pentru analiza acestora, una pentru caracteristicile categoriale È™i una pentru caracteristicile numerice
cat_cols = [col for col in dft.columns if dft[col].dtype == np.object]
num_cols = [col for col in dft.columns if dft[col].dtype != np.object]
```

```python
#afisarea cardinalitatii coloanelor categorice
cards = [len(dft[col].unique()) for col in cat_cols]
fig,ax = plt.subplots(figsize=(18,6))
sns.barplot(x=cat_cols, y=cards)
ax.set_xlabel('Feature')
```

```python
#distribuÈ›ia variabilelor categoriale
fig,ax = plt.subplots(figsize=(10,6))
print(dft['Sex'].value_counts() / len(dft))
sns.countplot(dft['Sex'])
ax.set_ylabel('Number of Patients')
ax.set_title('Number of patients by sex')
```

```python
#distribuÈ›ia colesterolului Ã®n funcÈ›ie de starea bolii
fig,ax = plt.subplots(figsize=(10,6))
data_train_disease = dft[dft['HeartDisease']==1]
data_train_no_disease = dft[dft['HeartDisease']==0]

sns.distplot(data_train_no_disease['Cholesterol'], label='No disease')
sns.distplot(data_train_disease['Cholesterol'], label='Disease')
ax.legend()
ax.set_title('Cholesterol Distribution')
```

```python
#check numÄƒrul vÃ¢rstelor
plt.figure(figsize=(25,12))
sns.set_context('notebook',font_scale = 1.5)
sns.barplot(x=df.Age.value_counts()[:10].index,y=df.Age.value_counts()[:10].values)
plt.tight_layout()
```

```python
#afiÈ™area relaÈ›ia dintre sex È™i slope
plt.figure(figsize=(18,9))
sns.set_context('notebook',font_scale = 1.5)
sns.countplot(df['Sex'],hue=df["ST_Slope"])
plt.tight_layout()
```

```python
#afiÈ™area relaÈ›iei dintre fucÈ›ii
plt.figure(figsize=(14, 14))
sns.heatmap(df.corr(), annot=True)
plt.show()
```

```python
#reprezentarea funcÈ›iilor aparte
fig = plt.figure(figsize=(20, 20))
ax = fig.gca()
df.plot(kind='density', subplots=True, layout=(4, 4), sharex=False, ax=ax)
plt.show()
```

```python
#diagrama cu case pentru fiecÄƒre funcÈ›ie.
fig = plt.figure(figsize=(20, 20))
ax = fig.gca()
df.plot(kind='box', subplots=True, layout=(4, 4), sharex=False, ax=ax)
plt.show()
```
